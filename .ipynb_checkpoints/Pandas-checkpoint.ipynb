```python
import pandas as pd
DFvuelos = pd.read_csv('airlines_flights_data.csv', index_col=0)
```

```python
print(DFvuelos)
print("\n  \n *************** Con .size cuenta todo registros por campos *****************************")
print(DFvuelos.size)
print("\n  \n *************** Con .shape cuenta solo registros e indica los campos *****************************")
print(DFvuelos.shape)
```

```python
# sacar datos como una serie
sPaises = DFvuelos['source_city']
print (type(sPaises))
print (sPaises.head())
# sacamos el listado de paises de origen pero repetidos, vamos a ver cuantos son (serie.size)
# y si queremos ver archivos unicos (serie_con_duplicados.drop_duplicates())
print(sPaises.size)   # 300153 elementos
sPaisesSinDuplicados = sPaises.drop_duplicates()
print(sPaisesSinDuplicados.size)
print(sPaisesSinDuplicados.values)
```

```python
# hacer consultas especificas
DFConsulta = DFvuelos[DFvuelos["destination_city"] == "Mumbai"]
print ("Encabezado consulta")
print (DFConsulta.head())

print ("\n \n")
print(" ********************************* Numero de registros ******************************")
print(DFConsulta.size)
print(DFConsulta.shape, "registros con ciudad Mumbai")
```

```python
DFConsulta = DFvuelos[ (DFvuelos["destination_city"] == "Delhi") & (DFvuelos["source_city"] == "Bangalore") & (DFvuelos["price"] < 3000) ]

print ("Encabezado consulta")
print (DFConsulta.head())

print ("\n \n")
print(" ********************************* Numero de registros ******************************")
print(DFConsulta.size)
print(DFConsulta.shape, "registros con ciudad Delhi")
```

```python
# tambien se puede usar df.query
DFConsulta = DFvuelos.query("departure_time == 'Night' and stops == 'one' and price > 3000 and price < 4000")

print ("Encabezado consulta")
print (DFConsulta.head())

print ("\n \n")
print(" ********************************* Numero de registros ******************************")
print(DFConsulta.size)
print(DFConsulta.shape, "registros ")
```

```python
# seleccionar columnas campos
DFConsulta = DFvuelos[["price", "source_city"]]

print ("Encabezado consulta")
print (DFConsulta.head())

print ("\n \n")
print(" ********************************* Numero de registros ******************************")
print(DFConsulta.size)
print(DFConsulta.shape, "registros ")
```

```python
# ahora vamos a filtrar y a seleccionar filas segun una condicion con  loc

DFConsulta = DFvuelos.loc[:, ['source_city', 'price']]  # : todos los registros

print ("Encabezado consulta")
print (DFConsulta.head())

print ("\n \n")
print(" ********************************* Numero total de registros ******************************")
print(DFConsulta.size)
print(DFConsulta.shape, "registros ")



DFConsulta = DFvuelos.loc[ (DFvuelos["price"]<2000) & (DFvuelos["stops"]=="one") , ['source_city', 'price']]  # : todos los registros

print ("Encabezado consulta")
print (DFConsulta.head())

print ("\n \n")
print(" ********************************* Numero total de registros ******************************")
print(DFConsulta.size)
print(DFConsulta.shape, "registros ")
```
