## Objetos:
### En Python todos son objetos las variables, las funciones y tienen asociados metodos
### ademas al momento de asignar a una variable lo que ya otra variable tiene  en relaidad se esta realizando  es una referencia al objeto, veamos el siguiente ejemplo:

```python
a=[1,2,3]
b= a
a.append(5)
print(" se refleja en cambio en b porque hacen referencia al mismo objeto" , b, "si lo quisiera dejar independiente tendria que declararlo con =[1,2,3]" )
```

## Referencias Dinamicas:
### En python no hay que declarar el tippo de variable como en otro lenguajes sino que de acuerdo al valor  o asignación se crea el tipo, ejemplo

```python
x=6
print (type(x))
x="foolih beat"
print (type(x))
# se le asigno a x un nuevo valor y acorde a este se modificó su tipo.
```

## Saber que tipo es un objeto:
#### podemos saber  de dos maneras, la primera es usando type(objeto) usada en el ejemploo anterior y la otra manera es usando 
#### la sentencia isinstance(objeto, tipo) pero de esta manera responde es un True o False si correnponde con el tipo dado
#### veamos un ejemplo


```python
a = 5; b = 4.5
isinstance(b, float)
```

```python
isinstance(b, (int, float))
```

## Verificar si una Variable hace Referencia al Mismo Objeto
Como se decia antes  en python todos son objetos y al igualar una variable con otra se esta haciendo referencia al mismo objeto, entonces para poder verificar esta referencia usamos la sentencia "is"


```python
print (a)
print (b)
c=a
print (c)
print( c is a)
print( a is not c)


```

## Objetos mutables e inmutables


```python
# como vemos  las cadenas de texto son inmutables, para transtomar una cadena podemos  nombrar otra cadena con lo transformado
cadena = "Esto es una cadena"
otraCadena = cadena.replace("cadena", "Larga Cadena")
print (cadena)
print(otraCadena)
```

```python
# con list se puede crear una lista de un string
listaCadena = list(cadena)
print(listaCadena)
```
