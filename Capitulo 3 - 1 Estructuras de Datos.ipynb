# Estructuras de Datos Integrados

#### Este capítulo aborda ciertas capacidades del lenguaje Python que emplearemos profusamente a lo largo del libro. Librerías adicionales, como pandas y NumPy, añaden funcionalidad computacional avanzada para grandes conjuntos de datos, pero están diseñadas para usarse junto con las herramientas de manipulación de datos integradas en Python. Empezaremos con las estructuras de datos esenciales de Python: tuplas, listas, diccionarios y conjuntos; después hablaremos de la creación de nuestras propias funciones de Python reutilizables y, por último, veremos la mecánica de los objetos archivo de Python y su interacción con el disco duro local del usuario.
