### Tuplas

#### Una tupla es una secuencia de objetos Python inmutable y de longitud fija que, una vez asignada, no puede modificarse. La forma más sencilla de crear una tupla es mediante una secuencia de valores separados por comas y encerrados entre paréntesis



```python
# Creacion de tupla

tup= (4,5,6)
print( "tipo: ", type(tup), " Contenido: ", tup)
```

```python
# tambien se puede crear un tupla sin parentesis, con solo separar los elementos por comas
tupla = 1, 2, 3
print( "tipo: ", type(tupla), " Contenido: ", tupla)
```

```python
# con la palabra tuple se puede convertir una lista en tupla y una cadena de texto en tupla
tup2 = tuple([7,8,9])
print( "tipo: ", type(tup2), " Contenido: ", tup2)
tup3 = tuple("Hola")
print( "tipo: ", type(tup3), " Contenido: ", tup3)
```

```python
# Si un objeto contenido en una tupla es mutable, como una lista por ejemplo, se puede modificar en su ubicación:
tup4 = ("A","B",[3,4])
tup4[2].append(5)
print (tup4)
```


## Listas:

### A diferencia de las tuplas, las listas tienen longitud variable y su contenido se puede modificar. Las listas son mutables, y pueden definirse utilizando paréntesis cuadrados [] o la función de tipo list

```python
a_list = [2, 3, 7, None]
print (a_list)

# hacer una lista basada en una tupla
tup = ("foo", "bar", "baz")
b_list = list(tup)
print ("Tipo Tup: ",type(tup), "tipo b_list: ", type(b_list))
#contenido de la tupla y la lista son iguales pero el tipo de dato es diferente
print ("Contenido Tup: ", tup, "Contenido b_list: ", b_list) 


# crear una lista a partir de una cadena de texto
c_list = list("Hello")
print ("Tipo c_list: ", type(c_list), "Contenido: ", c_list)
```

```python
# se pueden agregar valores a las listas utilizando el método append
c_list.append("World")
print ("Tipo c_list: ", type(c_list), "Contenido: ", c_list)
# se pueden agregar valores a las listas utilizando el método insert, 
# que permite especificar la ubicación del nuevo valor
c_list.insert(0, "Start")
print ("Tipo c_list: ", type(c_list), "Contenido: ", c_list)
```

```python
# tambien podemos eliminar elementos de una lista utilizando el método pop, que elimina 
# y devuelve el último 
# elemento de la lista por defecto, o el elemento en la ubicación especificada asi:
print ("c_list Contenido: ", c_list)
print("Elemento eliminado posicion 0: ", c_list.pop(0))
print ( "Contenido: ", c_list)

# se puede eliminar por valor utilizando el método remove, 
# que elimina la primera ocurrencia del valor especificado
c_list.remove("World")
print ( "Contenido: ", c_list)
```

```python
# con extend se pueden agregar varios elementos a una lista a la vez
print ("c_list Contenido: ", c_list)
c_list.extend(["End", "Finish"])
print ("Tipo c_list: ", type(c_list), "Contenido: ", c_list)
```


### Corte o rebanado de listas

#### Se pueden seleccionar partes de la mayoría de los tipos de secuencia empleando la notación de corte, que en su forma básica consiste en pasar start:stop  [start incluido : Stop excluido] al operador de indexado []:




```python
seq = [7, 2, 3, 7, 5, 6, 0, 1]
print(seq[1:5])

# si se omite el strat o el stop se asume el inicio o el final de la secuencia respectivamente
print("Lista Completa: ",seq)
print("primeras 5 posiciones: ",seq[:5])
print("toma desde posicion 3: ",seq[3:])
```



## Diccionarios:

#### Puede que el diccionario o dict sea la estructura de datos integrada de Python más importante de todas. En otros lenguajes de programación, a los diccionarios también se les llama mapas hash o arrays asociativos. Un diccionario almacena una colección de pares clave-valor, donde la clave y el valor son objetos Python. Cada clave está asociada a un valor, de modo que dicho valor se pueda recuperar, insertar, modificar o borrar convenientemente dada una determinada clave. Una forma de crear un diccionario es utilizando llaves {} y signos de dos puntos para separar claves y valores

```python
empty_dict = {}
d1 = {"a": "some value", "b": [1, 2, 3, 4]}
print (d1)

# puede acceder a elementos de un diccionario utilizando corchetes y la clave
print ("Valor de la clave a: ", d1["a"])
# se pueden modificar los valores de un diccionario asignando un nuevo valor 
# a una clave existente
d1["b"] = "otra cosa"   
print (d1)
# se pueden agregar nuevos pares clave-valor a un diccionario asignando 
# un valor a una nueva clave
d1["c"] = 123
print (d1)

# se pueden eliminar pares clave-valor de un diccionario utilizando la palabra clave del
# del seguida del nombre del diccionario y la clave a eliminar
del d1["a"]
print (d1)

# tambien se pueden eliminar pares clave-valor utilizando el método pop, que elimina 
# y devuelve el valor asociado a la clave especificada
print("Valor eliminado con .pop: ", d1.pop("b"))
```

#### Diccionarios - Metodo keys y values

##### El método keys y values proporciona iteradores de las claves y los valores del diccionario, respectivamente. El orden de las claves depende del orden de su inserción, y estas funciones dan como resultado las claves y los valores en el mismo orden respectivo

```python
# los diccionarios se pueden  sacar en listas basados en sus claves, 
# valores o pares clave-valor utilizando los métodos keys, values y items respectivamente
d1 = {"a": "some value", "b": [1, 2, 3, 4]}
print ("Claves: ", d1.keys())
print ("Valores: ", d1.values())
print ("Items: ", d1.items())

# se puede comprobar si una clave está presente en un diccionario utilizando 
# la palabra clave in
print ("a es una clave del diccionario d1? ", "a" in d1)

# con el metodo update se pueden actualizar los valores de un diccionario con los valores 
# de otro diccionario, o con pares clave-valor especificados como argumentos de palabra clave
d1.update({"a": "nuevo valor", "b": [5, 6, 7]})
print ("Diccionario actualizado: ", d1)
```


#### Diccionarios:

#### mapping = {}
#### for key, value in zip(key_list, value_list):
#### mapping[key] = value


```python
# Ejemplo
mapping = {}
mapping.update(zip("abcde", range(5)))  
print ("Diccionario creado a partir de secuencias: ", mapping)

# CON FOR
mapping = {}
for key, value in zip("FGHIJ", range(10,20,2)):
    mapping[key] = value
print ("Diccionario creado con bucle for: ", mapping)
```

## Funcion Zip

##### Características y usos clave:

#### Combinar secuencias: Si a = [1, 2] y b = ["uno", "dos"], zip(a, b) produce (1, "uno"), (2, "dos"). 
#### Longitudes desiguales: Se detiene automáticamente cuando se agota el iterable más corto.     
#### Uso en for loops:  Ideal para iterar sobre varias listas simultáneamente, por ejemplo: for x, y in zip(lista1, lista2):.     
#### Crear diccionarios: Se puede usar dict(zip(claves, valores)) para crear diccionarios de manera rápida.
#### Desempaquetar: Con el operador * (zip(*zip_objeto)), se pueden separar los elementos combinados de nuevo en sus estructuras originales


## Valores predeterminados

##### Es común tener lógica como la siguiente:
##### if key in some_dict:
##### value = some_dict[key]
##### else:
##### value = default_value

##### Así, los métodos de diccionario get y pop pueden tomar un valor predeterminado que devolverán, de forma que el bloque anterior if-else se podría escribir con más sencillez así:

##### value = some_dict.get(key, default_value)

##### De forma predeterminada, get devolverá None si la clave no está presente, mientras que pop producirá una excepción. Con valores de configuración, puede ser que los valores de un diccionario sean otro tipo de colección, como una lista. Por ejemplo, podríamos pensar en categorizar una lista de palabras por sus letras iniciales como un diccionario de listas:

```python
words = ["apple", "bat", "bar", "atom", "book","andres","armando"]

by_letter = {}
for word in words:
    letter = word[0]
    if letter not in by_letter:
        by_letter[letter] = [word]
    else:
        by_letter[letter].append(word)

print("Palabras agrupadas por letra inicial:")
for letter, wordsl in by_letter.items():
    print(f"{letter}: {wordsl}")
    
# este mismo codigo se podria escribir de la siguiente manera:
print ( "lista words: ", words)
by_letter = {}
for word in words:
    letter = word[0]
    print( word[0] )
    by_letter.setdefault(letter, []).append(word)
print("Palabras agrupadas por letra inicial: Optimizado")
for letter, wordsl in by_letter.items():
    print(f"{letter}: {wordsl}")
    
# el modulo collections de la biblioteca estándar de Python proporciona una 
# clase llamada defaultdict que hace que este tipo de agrupación sea aún 
# más fácil. Un defaultdict es como un diccionario normal, 
# pero si se intenta acceder a una clave que no existe, 
# en lugar de generar un error, se crea automáticamente 
# un nuevo valor utilizando una función de fábrica proporcionada.
# En este caso, podemos usar list como función de fábrica para crear 
# una nueva lista cada vez que se accede a una clave que no existe:
from collections import defaultdict

print ( "lista words: ", words)
by_letter = defaultdict(list)
for word in words:
    letter = word[0]
    by_letter[letter].append(word)
print("Palabras agrupadas por letra inicial: Usando defaultdict")
for letter, words in by_letter.items():
    print(f"{letter}: {words}")
    
```

### Tipos de claves de diccionario válidas
Aunque los valores de un diccionario pueden ser cualquier objeto
Python, generalmente las claves tienen que ser objetos inmutables, como
tipos escalares (int, float, string) o tuplas (todos los objetos de la tupla
tienen que ser también inmutables). Aquí el término técnico es hashability,
o capacidad de resumen. Se puede verificar si un objeto es resumible o
hashable (es decir, se puede usar como clave en un diccionario) con la
función hash:

```python

print (hash("string"))

print(hash((1, 2, (2, 3))))

print (hash((1, 2, [2, 3])) ) # falla porque las listas son mutables
```

## Conjunto o set
Un conjunto o set es una colección desordenada de elementos únicos. Se
pueden crear de dos maneras, mediante la función set o con un literal de
conjunto con llaves:

```python
# con set se crean conjuntos asi:
conjuntoConSet= set([1, 2, 3, 4, 5, 5, 5, 5, 5, 5, 1, 2, "a", "a", "A", "a", "a"])

# se pueden crear con llaves de conjunto tambien:
conjuntoLlaves= {1, 2, 3, 4, 5, 5, 5, 5, 5, 5, 1, 2}

print ("Conjunto con set:",conjuntoConSet)
print ("Conjunto con llaves:",conjuntoLlaves)


# union se puede hacer con el método union o con el operador |
union= conjuntoConSet.union(conjuntoLlaves)
print ("Union con metodo union: ", union)
union2= conjuntoConSet | conjuntoLlaves
print ("Union con operador |: ", union2)


# intersección se puede hacer con el método intersection o con el operador &
interseccion= conjuntoConSet.intersection(conjuntoLlaves)       
print ("Intersección con metodo intersection: ", interseccion)
interseccion2= conjuntoConSet & conjuntoLlaves
print ("Intersección con operador &: ", interseccion2)


# diferencia se puede hacer con el método difference o con el operador -
diferencia= conjuntoConSet.difference(conjuntoLlaves)
print ("Diferencia con metodo difference: ", diferencia)
diferencia2= conjuntoConSet - conjuntoLlaves
print ("Diferencia con operador -: ", diferencia2)  
```
